## Dropping constant features using VarianceThreshold

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vr = VarianceThreshold(0)
```

```python
vr
```

```python
vr.fit(data)
```

```python
vr.get_support()
```

```python
data.columns[vr.get_support() == False]
# drop these columns 
```

```python
data.columns[vr.get_support()]
# not dropped
```

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[vr.get_support()]]
# will droped
print(len(constant_columns))
```

```python
for i in constant_columns :
    print(i)
```

```python
data.drop(constant_columns , axis = 1 , inplace=True)
```

```python
def correlation(dataset , threshold) : 
    correlated_columns = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i) :
            if abs(corr_matrix.iloc[i,j] > threshold) :# we are interested in absolute coeff value
                col_name = corr_matrix.columns[i] # get the name of col 
                correlated_columns.add(col_name)
    return correlated_columns
```

```python
load_boston_correlation = correlation(X_train, 0.6)
load_boston_correlation
```

```python
train_corr = correlation(x_train, 0.8)
train_corr
```
